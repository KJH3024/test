### 전국 신용카드 데이터 불러오기

```python
import pandas as pd

card_df = pd.read_csv('./csv/지역_유형별 신용카드_(단위_억).csv', encoding= 'utf-8')
```

### 공백 제거 및 인덱스 변경

```python
# 공백 제거
card_df = card_df.replace(" ", "")

# 컬럼명 변경 및 인덱스 지정
card_df.rename(columns={'지역, 카테고리':'날짜'}, inplace= True)
card_df.set_index('날짜', inplace= True)

card_df.head()
```

### 지역 별로 데이터 분리

```python
# 지역 확인
name = card_df.filter(regex='_합계')
name.columns
```

```python
# 전국
card_all_df = card_df.filter(regex='전국_')

# 광역 시도
card_seoul_df = card_df.filter(regex='서울_')
card_busan_df = card_df.filter(regex='부산_')
card_deagu_df = card_df.filter(regex='대구_')
card_incheon_df = card_df.filter(regex='인천_')
card_gwangju_df = card_df.filter(regex='광주_')
card_daejeon_df = card_df.filter(regex='대전_')
card_ulsan_df = card_df.filter(regex='울산_')

# 도
card_gyeonggi_df = card_df.filter(regex='경기_')
card_gangwon_df = card_df.filter(regex='강원_')
card_chungcheong_df = pd.merge(card_df.filter(regex='충북_'), card_df.filter(regex='충남_'), on="날짜", how="left")
card_chungbook_df = card_df.filter(regex='충북')
card_chungnam_df = card_df.filter(regex='충남')
card_jeolla_df = pd.merge(card_df.filter(regex='전북_'), card_df.filter(regex='전남_'), on="날짜", how="left")
card_jeonbook_df = card_df.filter(regex='전북')
card_jeonnam_df = card_df.filter(regex='전남')
card_kyungsang_df = pd.merge(card_df.filter(regex='경북_'), card_df.filter(regex='경남_'), on="날짜", how="left")
card_kyungbook_df = card_df.filter(regex='경북')
card_kyungnam_df = card_df.filter(regex='경남')
card_jeju_df = card_df.filter(regex='제주_')
card_etc_df = card_df.filter(regex='기타_')
```

### csv 파일로 저장

```python
# 전국
card_all_df.to_csv('./data/카드_전국.csv', encoding='utf-8', index=True)

# 광역 시도
card_seoul_df.to_csv('./data/카드_서울.csv', encoding='utf-8', index=True)

card_busan_df.to_csv('./data/카드_부산.csv', encoding='utf-8', index=True)
card_deagu_df.to_csv('./data/카드_대구.csv', encoding='utf-8', index=True)
card_incheon_df.to_csv('./data/카드_인천.csv', encoding='utf-8', index=True)
card_gwangju_df.to_csv('./data/카드_광주.csv', encoding='utf-8', index=True)
card_daejeon_df.to_csv('./data/카드_대전.csv', encoding='utf-8', index=True)
card_ulsan_df.to_csv('./data/카드_울산.csv', encoding='utf-8', index=True)

# 도
card_gyeonggi_df.to_csv('./data/카드_경기.csv', encoding='utf-8', index=True)
card_gangwon_df.to_csv('./data/카드_강원.csv', encoding='utf-8', index=True)
card_chungcheong_df.to_csv('./data/카드_충청.csv', encoding='utf-8', index=True)
card_chungbook_df.to_csv('./data/카드_충북.csv', encoding='utf-8', index=True)
card_chungnam_df.to_csv('./data/카드_충남.csv', encoding='utf-8', index=True)
card_jeolla_df.to_csv('./data/차드_전라.csv', encoding='utf-8', index=True)
card_jeonbook_df.to_csv('./data/카드_전북.csv', encoding='utf-8', index=True)
card_jeonnam_df.to_csv('./data/카드_전남.csv', encoding='utf-8', index=True)
card_kyungsang_df.to_csv('./data/카드_경상.csv', encoding='utf-8', index=True)
card_kyungbook_df.to_csv('./data/카드_경북.csv', encoding='utf-8', index=True)
card_kyungnam_df.to_csv('./data/카드_경남.csv', encoding='utf-8', index=True)
card_jeju_df.to_csv('./data/카드_제주.csv', encoding='utf-8', index=True)
card_etc_df.to_csv('./data/카드_기타.csv', encoding='utf-8', index=True)
```
